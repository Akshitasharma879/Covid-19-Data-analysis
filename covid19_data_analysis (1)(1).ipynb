```python
vaccine_df = vaccine_df.drop(columns=['Sputnik V (Doses Administered)', 'AEFI' , '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'], axis = 1)
```

```python
 # Forward filling for cumulative vaccination data
    
cumulative_cols = ['Total Doses Administered', 'Total', 
                    'First Dose Administered', 'Second Dose Administered']
vaccine_df[cumulative_cols] = vaccine_df[cumulative_cols].fillna(method='ffill')

```

```python
# Group-wise forward fill 

vaccine_df[cumulative_cols] = vaccine_df.groupby('State', group_keys=False)[cumulative_cols].apply(lambda x: x.fillna(method='ffill'))
```

```python
# Fill remaining NaN with 0
vaccine_df.fillna(0, inplace=True)
```

```python
# Interpolation for demographic data
demographic_cols = ['Male (Doses Administered)', 'Female (Doses Administered)']

vaccine_df[demographic_cols] = vaccine_df[demographic_cols].interpolate(method='linear')
```

```python

# Grouping data by state and summing up male and female vaccinations
statewise_vaccination = vaccine_df.groupby('State')[['Male (Doses Administered)', 'Female (Doses Administered)']].sum().reset_index()

# Selecting top 10 states with highest vaccinations
top_10_states = statewise_vaccination.nlargest(10, ['Male (Doses Administered)', 'Female (Doses Administered)'])

# Reshaping data for clustered bar chart
top_10_vaccination_melted = top_10_states.melt(id_vars='State', 
                                                var_name='Gender', 
                                                value_name='Doses')

# Setting figure size
plt.figure(figsize=(14, 8))

# Creating a clustered bar chart
sns.barplot(data=top_10_vaccination_melted, 
            x='State', 
            y='Doses', 
            hue='Gender', 
            palette='Set2')

# Adding titles and labels
plt.title('Top 10 States - Male & Female COVID-19 Vaccination', fontsize=16)
plt.xlabel('States', fontsize=14)
plt.ylabel('Doses Administered', fontsize=14)
plt.xticks(rotation=90)  # Rotating x-axis labels for better readability
plt.legend(title='Gender')

# Display the plot
plt.show()
```

```python
# Grouping by state for vaccination total
statewise_data = vaccine_df.groupby("State")[["Total"]].sum().reset_index()

# Grouping by state for COVID data
active_cases_data = covid_df.groupby("State/UnionTerritory")[["Cured", "Deaths", "Active_cases"]].max().reset_index()

# Merge datasets
merged_statewise = pd.merge(statewise_data, active_cases_data, 
                            left_on="State", right_on="State/UnionTerritory", how="inner")
# Calculate recovery rate safely
merged_statewise["Recovery_Rate"] = merged_statewise["Cured"] / (
    merged_statewise["Cured"] + merged_statewise["Deaths"]
)

# Normalize bubble size for visualization
max_active = merged_statewise["Active_cases"].max()
merged_statewise["Bubble_Size"] = (merged_statewise["Active_cases"] / max_active) * 1000 + 100
```

```python
# Normalize bubble size
max_active = merged_statewise["Active_cases"].max()
merged_statewise["Bubble_Size"] = (merged_statewise["Active_cases"] / max_active) * 1000 + 100

# Create the scatter plot
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    x=merged_statewise["Total"],
    y=merged_statewise["Recovery_Rate"],
    s=merged_statewise["Bubble_Size"],
    alpha=0.7,
    c=merged_statewise["Active_cases"],
    cmap="plasma",  #  Using a vibrant colormap
    edgecolors="black"
)

# Annotate high active cases
for i, row in merged_statewise.iterrows():
    if row["Active_cases"] > 50000:
        plt.annotate(row["State"], (row["Total"], row["Recovery_Rate"]), fontsize=9)

# Chart labels and title
plt.xlabel("Total Vaccinations", fontsize=12)
plt.ylabel("Recovery Rate", fontsize=12)
plt.title("Vaccination vs Recovery Rate (Bubble Size = Active Cases)", fontsize=16)
plt.ylim(0.94, 1.01)

# Add color bar to show active cases gradient
cbar = plt.colorbar(scatter)
cbar.set_label("Active COVID-19 Cases")

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
